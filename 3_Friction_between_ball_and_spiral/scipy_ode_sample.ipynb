```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def pend(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt
```

```python
y0 = [np.pi - 0.1, 0.0]
t = np.linspace(0, 10, 101)

b = 0.25
c = 5.0
```

```python
from scipy.integrate import odeint
sol = odeint(pend, y0, t, args=(b, c))
```

```python
fig, ax = plt.subplots()
ax.plot(t, sol)
ax.legend({"y", "y'"})
```
